```python
def get_assignments(line):
    assignments = []
    split = line.split(",")
    for e in split:
        numbers = e.split("-")
        assignments.append([int(numbers[0]), int(numbers[1])])
    return assignments

def check_subgroup(assignments):
    if (assignments[0][0] <= assignments[1][0]) & (assignments[0][1] >= assignments[1][1]):
        return True
    elif (assignments[1][0] <= assignments[0][0]) & (assignments[1][1] >= assignments[0][1]):
        return True
    else:
        return False

# [[16, 66], [13, 14]]
def check_overlap(assignments):
    print (assignments)

    assignment_1 = list(range(assignments[0][0], assignments[0][1]+1))
    assignment_2 = list(range(assignments[1][0], assignments[1][1]+1))

    print(assignment_1)
    print(assignment_2)

    for i in assignment_1:
        if i in assignment_2:
            return True
    
    for i in assignment_2:
        if i in assignment_1:
            return True
    
    return False
```

```python
# 23-33,24-65

with open("input.txt") as f:
    lines = []
    for line in f.readlines():
        lines.append(line.replace("\n", ""))
```

```python
subgroup = 0
overlap = 0
for line in lines:
    assignments = get_assignments(line)
    print(assignments)
    if (check_subgroup(assignments)):
        subgroup += 1
        print("subgroup")
    if (check_overlap(assignments)):
        overlap += 1
        print("overlap")

print("subgroup: ", subgroup) #first answer 595 was correct
print("overlap: ", overlap) #1000 was false, 743 was false, 952 was right. the ranges exclude the last element by default.
```
